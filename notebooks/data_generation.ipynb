# Data generation for Rust tests

```python
from sklearn.datasets import make_blobs
from gaussian_mixtures import *
from pathlib import Path

data, z = make_blobs(n_samples=10000, centers=10, n_features=2, random_state=7)  # pylint: disable=W0632
# responsibilities = Likelihood(np.random.dirichlet(10 * [1.0], data.shape[0]))
responsibilities = np.zeros((10000, 10))
responsibilities[np.arange(10000),z] = 1.0
gmm = GaussianMixtureModel(np.zeros(0), np.zeros(0), np.zeros(0))
maximize_(gmm, responsibilities, data)

print(gmm.means)
print(gmm.covs)

path = Path('../rust/data')
np.save(path / "data.npy", data)
np.save(path / "responsibilities.npy", responsibilities)
np.save(path / "means.npy", gmm.means)
np.save(path / "weights.npy", gmm.weights)
np.save(path / "covs.npy", gmm.covs)

print("habemus data")
```
